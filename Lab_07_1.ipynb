```python
# Imports
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
# For reproducibility
torch.manual_seed(1)
```

```python
# Training and Test Dataset
x_train = torch.FloatTensor([[1, 2, 1],
                             [1, 3, 2],
                             [1, 3, 4],
                             [1, 5, 5],
                             [1, 7, 5],
                             [1, 2, 5],
                             [1, 6, 6],
                             [1, 7, 7]
                            ])                                    # m * 3
y_train = torch.LongTensor([2, 2, 2, 1, 1, 1, 0, 0])              # m

x_test = torch.FloatTensor([[2, 1, 1], [3, 1, 2], [3, 3, 4]])     # m' * 3
y_test = torch.LongTensor([2, 2, 2])                              # m'
```

```python
# Model
class SoftmaxClassifierModel(nn.Module):
  def __init__(self):
    super().__init__()
    self.linear = nn.Linear(3, 3)         # 3 * 3
  def forward(self, x):
    return self.linear(x)                 # m * 3
```

```python
model = SoftmaxClassifierModel()
```

```python
# optimizer 설정
optimizer = optim.SGD(model.parameters(), lr=0.1)
```

```python
# Training function
# train 하는 함수를 만들자
def train(model, optimizer, x_trian, y_train):
  nb_epochs = 20
  for epoch in range(nb_epochs + 1):

    # H(x) 계산 (use prediction instead of hypothesis)
    prediction = model(x_train)

    # Cost 계산
    cost = F.cross_entropy(prediction, y_train)

    # cost 로 H(x) 개선
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    print('Epoch {:4d}/{} Cost: {:.6f}'.format(
        epoch, nb_epochs, cost.item()
    ))
```

```python
# Test function(Validation)
# test 하는 함수를 만들자
def test(model, optimizer, x_test, y_test):
  prediction = model(x_test)          # size = m' * 3 (x_test 의 size 가 m' * 3 이므로)
  predicted_classes = prediction.max(1)[1]
  correct_count = (predicted_classes == y_test).sum().item()
  cost = F.cross_entropy(prediction, y_test)

  print('Accuracy: {}% Cost: {:.6f}'.format(
      correct_count / len(y_test) * 100, cost.item()
  ))
```

```python
# Run
train(model, optimizer, x_train, y_train)       # model equals: theta
```

```python
test(model, optimizer, x_test, y_test)
```

# Learning Rate
learning rate 가 너무 크면

대입식 θ:= θ - α∇(loss(x,θ)) 에서

과도하게 학습을 해 오히려 ∇(loss(x,θ)) 의 크기가 커지게 되고

다음 optimization 에서 더 커진 ∇(loss(x,θ)) 에 의해 α∇(loss(x,θ)) 값이 더 커지게 되면서

진동과 발산을 하면서 cost 의 값이 점점 커진다.

반대로 learning rate 가 너무 작으면

θ 가 가장 손실이 적은 θ 로 가기 전에 수렴하게 된다.

```python
# learning rate 가 너무 큰 경우
```

```python
model = SoftmaxClassifierModel()
```

```python
optimizer = optim.SGD(model.parameters(), lr=1e5)
```

```python
train(model, optimizer, x_train, y_train)
```

```python
# Data Preprocessing (데이터 전처리)
# Linear Regression data
x_train = torch.FloatTensor([[73, 80, 75],
                             [93, 88, 93],
                             [89, 91, 90],
                             [96, 98, 100],
                             [73, 66, 70]])                           # m * 3
y_train = torch.FloatTensor([[152], [185], [180], [196], [142]])      # m

mu = x_train.mean(dim=0)
sigma = x_train.std(dim=0)              # sigma: standard deviation,  .std(): standardization
norm_x_train = (x_train - mu) / sigma
print(norm_x_train)
```

```python
# Normalize 와 zero center 한 x 로 학습해서 성능확인
# Training with Preprocessed Data
class MultivariateLinearRegressionModel(nn.Module):
  def __init__(self):
    super().__init__()
    self.linear = nn.Linear(3, 1)   # 3 * 1

  def forward(self, x):
    return self.linear(x)

model = MultivariateLinearRegressionModel()

optimizer = optim.SGD(model.parameters(), lr=1e-5)
```

```python
def train(model, optimizer, x_trian, y_train):
  nb_epochs = 20
  for epoch in range(nb_epochs + 1):

    # H(x) 계산
    prediction = model(x_train)                   # (m, 3) * (3, 1) -> (m, 1)

    # Cost 계산
    cost = F.mse_loss(prediction, y_train)        # use torch.nn.functional

    # cost 로 H(x) 개선
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    print('Epoch {:4d}/{} Cost: {:.6f}'.format(
        epoch, nb_epochs, cost.item()
    ))
```

```python
train(model, optimizer, norm_x_train, y_train)
```
