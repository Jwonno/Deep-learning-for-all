```python
# Imports
import torch
import torch.nn as nn
import torch.optim as optim
from matplotlib import pyplot as plt
```

```python
# Data
X = torch.FloatTensor([[0, 0], [0, 1], [1, 0], [1, 1]])
Y = torch.FloatTensor([[0], [1], [1], [0]])
```

```python
# Linear Regression
class LinearRegressor(nn.Module):
  def __init__(self):
    super().__init__()
    self.fc = nn.Linear(1, 1, bias=True)    # y = w * x + b

  def forward(self, x):
    y = self.fc(x)

    return y
```

```python
model = LinearRegressor()
```

```python
# Loss Function using 'Mean Square Error'
criterion = nn.MSELoss()
```

```python
# optimizer 설정
optimizer = optim.SGD(model.parameters(), lr=1e-3)
```

```python
loss_stack = []                 # 매 epoch 마다 손실 함수값을 저장하기 위해 빈 리스트로 생성(그래프를 그리기위한 용)
for epoch in range(1001):

  # H(x) 계산
  y_hat = model(x)

  # Cost 계산
  loss = criterion(y_hat, y)

  # cost 로 H(X) 개선
  optimizer.zero_grad()         # 매 epoch 마다 누적된 계산을 초기화
  loss.backward()
  optimizer.step()

  loss_stack.append(loss.item())  # 그래프를 그리기 위해 손실 함수값(.item())만 리스트에 저장함

  # 100 번 마다 로그 출력
  if epoch % 100 == 0:
    print('Epoch: {} Loss: {}'.format(
        epoch, loss.item()
    ))
```

```python
# Prediction
# 예측은 학습하는 것이 아니기 때문에 with torch.no_grad(): 를 통해 requires_grad 를 비활성화한다.
with torch.no_grad():
  y_hat = model(x)

  # plot
  plt.figure(figsize=(10, 5))
  plt.subplot(121)
  plt.plot(loss_stack)
  plt.title("Loss")
  plt.subplot(122)
  plt.plot(x, y, '.b')
  plt.plot(x, y_hat, 'r-')
  plt.legend(['ground truth', 'prediction'])
  plt.title("Prediction")
  plt.show()
```
