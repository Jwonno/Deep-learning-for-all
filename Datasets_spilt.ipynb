#**Datasets**
*  Training data (80%)

*  Test data (10%~20%)

*  Validation data(0~10%)

```python
# Random Sampling
# sklearn 라이브러리의 train_set_split 을 이용
import numpy as np
from sklearn.model_selection import train_test_split

# 학습 데이터를 70%, 평가 데이터를 30% 로 비율 설정
X, y = np.arange(1000).reshape((100, 10)), np.arange(100)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
print('X_train:',X_train.size, ' X_test:', X_test.size)
print('y_train:', y_train.size, ' y_test:', y_test.size)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)        # train_data:test_data = 6:4
X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, test_size=0.5, random_state=1)  # test_data:validation_data = 5:5 , so train_data:test_data:validation_data = 6:2;2
print('X_train:',X_train.size, ' X_test:', X_test.size, ' X_val:',X_val.size)
print('y_train:', y_train.size, ' y_test:', y_test.size, ' y_val:',y_val.size)
```
