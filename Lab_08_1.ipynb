```python
import torch
import torch.optim as optim
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'

# for reproducibility
torch.manual_seed(1)
if device == 'cuda':
  torch.cuda.manual_seed_all(1)
```

```python
# XOR
X = torch.FloatTensor([[0, 0], [0, 1], [1, 0], [1, 1]]).to(device)
Y = torch.FloatTensor([[0], [1], [1], [0]]).to(device)
```

```python
linear = torch.nn.Linear(2, 1, bias=True)
sigmoid = torch.nn.Sigmoid()
```

```python
model = torch.nn.Sequential(linear, sigmoid).to(device)
```

```python
# define cost/loss & optimizer
criterion = torch.nn.BCELoss().to(device)
optimizer = optim.SGD(model.parameters(), lr=1)
```

```python
for step in range(1000 + 1):

  # H(X) 계산
  hypothesis = model(X)

  # Cost 계산
  cost = criterion(hypothesis, Y)

  # cost 로 H(X) 개선
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  # 매 100 번 마다 로그 출력
  if step % 100 == 0:
    print(step, cost.item())
```

```python
# learning rate 를 수정해보면? lr=1 -> lr=0.1
optimizer = optim.SGD(model.parameters(), lr=0.1)

for step in range(1000 + 1):

  # H(X) 계산
  hypothesis = model(X)

  # Cost 계산
  cost = criterion(hypothesis, Y)

  # cost 로 H(X) 개선
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  # 매 100 번 마다 로그 출력
  if step % 100 == 0:
    print(step, cost.item())
```

still cannot lower cost/loss

```python
# Accuracy computation
# True if hypothesis>0.5 else Fasle       예측한 값이 0.5 보다 크면 True, 작으면 False 로
with torch.no_grad():
  hypothesis = model(X)
  predicted = (hypothesis>0.5).float()
  accuracy = (predicted == Y).float().mean()
  print('\nHypothesis: {}\nCost: {}\nAccuracy: {}'.format(
      hypothesis.detach().cpu().numpy(),
      predicted.detach().cpu().numpy(),
      accuracy.item()
  ))
```
